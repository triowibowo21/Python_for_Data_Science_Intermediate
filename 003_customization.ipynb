```python
import matplotlib.pyplot as plt
year = [1950, 1951, 1952, 2100]
pop = [2.538, 2.57, 2.62, 10.85]

plt.plot(year, pop)
plt.show()
```

```python
plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('populasi')
plt.title('World Population Projection')
plt.yticks([0, 2, 4, 6, 8, 10]) #interval sumbu y
plt.show
```

```python
plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('populasi')
plt.title('World Population Projection')
plt.yticks([0, 2, 4, 6, 8, 10], 
        ['0', '2B', '4B', '6B', '8B', '100B']) #interval sumbu y dengan ket.bilion
plt.show
```

```python
year1 = [1800, 1850, 1900] + year
pop1 = [1.0, 1.262, 1.650] + pop
```

```python
plt.plot(year1, pop1)
plt.xlabel('Year')
plt.ylabel('populasi')
plt.title('World Population Projection')
plt.yticks([0, 2, 4, 6, 8, 10], 
        ['0', '2B', '4B', '6B', '8B', '100B']) #interval sumbu y dengan ket.bilion
plt.show
```
